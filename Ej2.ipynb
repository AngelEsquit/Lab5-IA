## Preguntas de discusión y respuestas

**1. ¿Puedes dibujar paredes para que Greedy produzca un camino mayor que BFS?**

Sí. Al construir muros largos en dirección al objetivo y que tienen un tope al final, Greedy insistirá en acercarse por su heurística local y terminará con un rodeo tardío y costoso. BFS, sin embargo, encontrará la ruta con menor número de pasos sin dejarse engañar.

**2. Muestra un escenario donde Greedy tome una ruta ineficiente por seguir la línea recta.**

En este caso son dos U juntas las que componen el camino.

**3. ¿Es Greedy óptimo? ¿Por qué?**

No es óptimo en general. Greedy usa solo la heurística (distancia estimada al objetivo) sin incorporar el costo real acumulado del camino recorrido, lo que lo hace vulnerable a obstáculos estratégicos.

---

**4. ¿Cuál algoritmo suele ser más rápido y cuál explora mayor área?**

- **Greedy:** Suele ser más rápido en tiempo de ejecución en muchos mapas simples por explorar menos en comparación inicial, pero sacrifica la calidad de la solución.
- **BFS:** Normalmente explora más nodos/área (es más costoso), pero garantiza la ruta más corta en número de pasos para grids sin pesos.
