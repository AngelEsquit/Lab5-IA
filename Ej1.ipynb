# Ejercicio 1 - El Laberinto de DFS y BFS

## Objetivo
Comparar BFS y DFS en un laberinto generado con Simple Random Maze.

```python
from IPython.display import display, Image, Markdown

imagenes = [
    ("img/ej1_bfs.png", "Breadth-First Search (BFS)"),
    ("img/ej1_dfs.png", "Depth-First Search (DFS)")
]

for img_path, titulo in imagenes:
    try:
        display(Markdown(f"### {titulo}"))
        display(Image(filename=img_path))
    except Exception as e:
        print(f"No se pudo cargar {img_path}: {e}")
```

## Preguntas de discusión y respuestas

**1. ¿Por qué BFS garantiza el camino más corto en este grid y DFS no?**

Porque BFS va revisando por niveles. Entonces, en un mapa sin pesos, cuando llega a la meta por primera vez, normalmente ya encontró el camino con menos pasos. En cambio DFS se mete por un camino hasta el fondo y puede encontrar una solución más larga antes que la mejor.

---

**2. Menciona una desventaja crítica de memoria para BFS y una de optimalidad para DFS.**

- **BFS:** gasta bastante memoria porque guarda muchos nodos al mismo tiempo.
- **DFS:** no siempre da el mejor camino, aunque encuentre una solución rápida en algunos casos.

---

**3. ¿Cuál es la diferencia visual en la frontera de expansión de ambos?**

- **BFS:** se ve como una expansión más pareja, como una ola con epicentro en el nodo inicial.
- **DFS:** se ve más como trazos largos hacia una dirección y luego regresos cuando se topa con bloqueos.

---

**4. ¿Qué ocurre si analizamos BFS y DFS en un laberinto Recursive Division?**

Los dos llegaron a la ruta más corta, y DFS incluso exploró menos área. Yo lo interpreto como que este tipo de laberinto le favoreció a DFS, porque encontró una buena rama relativamente rápido.

Entonces, en este caso no se ve una ineficiencia clara de DFS. La diferencia depende bastante de cómo esté armado el laberinto y del camino que le toque explorar primero a cada algoritmo.
